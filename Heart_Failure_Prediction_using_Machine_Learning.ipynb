```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Creating an instance of LabelEncoder
labelencoder = LabelEncoder()
```

```python
# converting non-numerical data to numerical data
heart_dataset['Sex']= labelencoder.fit_transform(heart_dataset['Sex'])
heart_dataset['ChestPainType']= labelencoder.fit_transform(heart_dataset['ChestPainType'])
heart_dataset['RestingECG']= labelencoder.fit_transform(heart_dataset['RestingECG'])
heart_dataset['ExerciseAngina']= labelencoder.fit_transform(heart_dataset['ExerciseAngina'])
heart_dataset['ST_Slope']= labelencoder.fit_transform(heart_dataset['ST_Slope'])
```

```python
# Display dataset
heart_dataset.head()
```

```python
x = heart_dataset.drop(['HeartDisease'],axis = 1)
```

```python
y =heart_dataset['HeartDisease']
```

```python
x.shape
```

```python
y.shape
```

## Splitting Training and testing Data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

## Standardize the Variables

```python
# Normalize the feature
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

# Using Logistic Regression model

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel.fit(x_train,y_train)
```

```python
y_pred = logmodel.predict(x_test)
```

```python
y_pred
```

##Evaluation

```python
 from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
 from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
 from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```

## using Support Vector Classifier

```python
from sklearn.svm import SVC
classifier = SVC()
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
```

## Predictions and Evaluations

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
```

```python
print(cm)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

##Using Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
```

## Predictions and Evaluations

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(y_test, y_pred)
```

```python
print(cm)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```
